19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
set_katarsis = {"Lukas Radzevičius", "Jokūbas Andriulis", "Alanas Brasas", "Emilija Kandratavičiūtė"}
set_akli = {"Kristijonas Valančius", "Benas Kalinauskas", "Jokūbas Andriulis"}
uni_set = set_katarsis.union(set_akli)
diff_set = set_katarsis.difference(set_akli)
intersect_set = set_katarsis.intersection(set_akli)
sym_diff_set = set_katarsis.symmetric_difference(set_akli)
print (uni_set)
print (diff_set)
print (intersect_set)
print (sym_diff_set)
```
